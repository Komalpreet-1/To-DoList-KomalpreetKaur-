```python
import tkinter as tk
from tkinter import messagebox
import json

class TodoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List Application")
        self.root.geometry("400x500")
        
        # Initialize an empty list to hold tasks
        self.tasks = []
        self.load_tasks()  # Load tasks from file on startup

        # Create UI components
        self.create_widgets()

    def create_widgets(self):
        # Label for the task entry
        self.task_label = tk.Label(self.root, text="Enter a new task:")
        self.task_label.pack(pady=10)

        # Entry for task input
        self.task_entry = tk.Entry(self.root, font=("Arial", 14))
        self.task_entry.pack(pady=10)

        # Add task button
        self.add_task_button = tk.Button(self.root, text="Add Task", command=self.add_task, width=15)
        self.add_task_button.pack(pady=5)

        # Mark task as completed button
        self.mark_task_button = tk.Button(self.root, text="Mark as Completed", command=self.mark_task_completed, width=15)
        self.mark_task_button.pack(pady=5)

        # Remove task button
        self.remove_task_button = tk.Button(self.root, text="Remove Task", command=self.remove_task, width=15)
        self.remove_task_button.pack(pady=5)

        # Listbox to display tasks
        self.task_listbox = tk.Listbox(self.root, font=("Arial", 12), selectmode=tk.SINGLE)
        self.task_listbox.pack(expand=True, fill='both', pady=10)

        # Update the Listbox with current tasks
        self.update_task_listbox()

    def add_task(self):
        """Add a task to the list."""
        task = self.task_entry.get().strip()
        if not task:
            messagebox.showerror("Input Error", "Please enter a task.")
            return
        self.tasks.append({"task": task, "completed": False})
        self.update_task_listbox()
        self.save_tasks()  # Save tasks to file
        self.task_entry.delete(0, tk.END)  # Clear input field

    def mark_task_completed(self):
        """Mark the selected task as completed."""
        try:
            selected_index = self.task_listbox.curselection()[0]
            self.tasks[selected_index]["completed"] = True
            self.update_task_listbox()
            self.save_tasks()
        except IndexError:
            messagebox.showerror("Selection Error", "Please select a task to mark as completed.")

    def remove_task(self):
        """Remove the selected task from the list."""
        try:
            selected_index = self.task_listbox.curselection()[0]
            del self.tasks[selected_index]
            self.update_task_listbox()
            self.save_tasks()  # Save tasks to file
        except IndexError:
            messagebox.showerror("Selection Error", "Please select a task to remove.")

    def update_task_listbox(self):
        """Update the Listbox to reflect the current tasks."""
        self.task_listbox.delete(0, tk.END)  # Clear current list
        for task in self.tasks:
            display_text = task["task"] + (" âœ”" if task["completed"] else "")
            self.task_listbox.insert(tk.END, display_text)

    def save_tasks(self):
        """Save the current tasks to a file."""
        with open("tasks.json", "w") as f:
            json.dump(self.tasks, f)

    def load_tasks(self):
        """Load tasks from a file at startup."""
        try:
            with open("tasks.json", "r") as f:
                self.tasks = json.load(f)
        except (FileNotFoundError, json.JSONDecodeError):
            self.tasks = []

# Create the root Tkinter window
root = tk.Tk()

# Create an instance of the TodoApp
todo_app = TodoApp(root)

# Start the Tkinter main loop
root.mainloop()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate some sample data for different charts
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic styling function for uniform plots
def style_plot(ax, title):
    ax.set_title(title)
    ax.grid(True)

# 1. Line Plot
def line_plot():
    fig, ax = plt.subplots()
    ax.plot(x, y, label='sin(x)')
    style_plot(ax, "Line Plot")
    ax.legend()
    plt.show()

# 2. Scatter Plot
def scatter_plot():
    fig, ax = plt.subplots()
    ax.scatter(x, y, color='purple', label='Data points')
    style_plot(ax, "Scatter Plot")
    ax.legend()
    plt.show()

# 3. Bar Chart
def bar_chart():
    categories = ['A', 'B', 'C', 'D', 'E']
    values = np.random.randint(1, 10, size=len(categories))
    fig, ax = plt.subplots()
    ax.bar(categories, values, color='orange')
    style_plot(ax, "Bar Chart")
    plt.show()

# 4. Histogram
def histogram():
    data = np.random.randn(1000)
    fig, ax = plt.subplots()
    ax.hist(data, bins=30, color='skyblue', edgecolor='black')
    style_plot(ax, "Histogram")
    plt.show()

# 5. Pie Chart
def pie_chart():
    labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
    sizes = [15, 30, 45, 10]
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    ax.set_title("Pie Chart")
    plt.show()

# 6. Box Plot
def box_plot():
    data = [np.random.rand(50), np.random.rand(50)]
    fig, ax = plt.subplots()
    ax.boxplot(data, vert=True, patch_artist=True)
    style_plot(ax, "Box Plot")
    plt.show()

# 7. Violin Plot
def violin_plot():
    data = [np.random.rand(100), np.random.rand(100) + 1]
    fig, ax = plt.subplots()
    ax.violinplot(data, showmeans=True)
    style_plot(ax, "Violin Plot")
    plt.show()

# 8. Heatmap
def heatmap():
    data = np.random.rand(10, 10)
    fig, ax = plt.subplots()
    heatmap = ax.imshow(data, cmap='viridis')
    fig.colorbar(heatmap)
    style_plot(ax, "Heatmap")
    plt.show()

# 9. Polar Plot
def polar_plot():
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(x, y)
    style_plot(ax, "Polar Plot")
    plt.show()

# 10. Area Plot (Stacked Plot)
def area_plot():
    df = pd.DataFrame({
        'A': np.random.rand(10),
        'B': np.random.rand(10),
        'C': np.random.rand(10),
    })
    fig, ax = plt.subplots()
    df.plot.area(ax=ax)
    style_plot(ax, "Area Plot (Stacked)")
    plt.show()

# Execute each plot function
line_plot()
scatter_plot()
bar_chart()
histogram()
pie_chart()
box_plot()
violin_plot()
heatmap()
polar_plot()
area_plot()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
import seaborn as sns

# Generate some random geographic data (latitude, longitude, and "intensity")
np.random.seed(42)
num_points = 1000
lats = np.random.uniform(-90, 90, num_points)      # Random latitudes
lons = np.random.uniform(-180, 180, num_points)    # Random longitudes
intensities = np.random.rand(num_points) * 100     # Random intensity values

# 1. Initialize the Map
plt.figure(figsize=(14, 8))
m = Basemap(projection='merc', llcrnrlat=-60, urcrnrlat=80,
            llcrnrlon=-180, urcrnrlon=180, resolution='c')

m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='lightgray', lake_color='aqua')

# 2. Convert lat/lon to map projection coordinates
x, y = m(lons, lats)

# 3. Create a 2D histogram to map intensity onto the geographic map
# Define grid size and apply histogram
heatmap, xedges, yedges = np.histogram2d(x, y, bins=100, weights=intensities)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# 4. Plot heatmap
plt.imshow(heatmap.T, extent=extent, origin='lower', cmap='hot', alpha=0.6)

# Add color bar
plt.colorbar(label="Intensity")

plt.title("Geographic Data Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# 1. Engagement Over Time: Plot Likes, Comments, Shares
def engagement_over_time(df):
    plt.figure(figsize=(12, 6))
    df[['likes', 'comments', 'shares']].plot()
    plt.title("Social Media Engagement Over Time")
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.legend(["Likes", "Comments", "Shares"])
    plt.grid(True)
    plt.show()

# 2. Follower Growth Over Time
def follower_growth(df):
    plt.figure(figsize=(12, 6))
    df['followers'].plot(color='darkgreen')
    plt.title("Follower Growth Over Time")
    plt.xlabel("Date")
    plt.ylabel("Followers")
    plt.grid(True)
    plt.show()

# 3. Daily Engagement Summary (Likes, Comments, Shares)
def engagement_distribution(df):
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[['likes', 'comments', 'shares']])
    plt.title("Distribution of Daily Engagement (Likes, Comments, Shares)")
    plt.ylabel("Count")
    plt.show()

# 4. Correlation Heatmap Between Metrics
def correlation_heatmap(df):
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title("Correlation Between Social Media Metrics")
    plt.show()

# 5. Likes and Comments Scatter Plot (Engagement Correlation)
def engagement_correlation(df):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='likes', y='comments', data=df, alpha=0.7)
    plt.title("Correlation Between Likes and Comments")
    plt.xlabel("Likes")
    plt.ylabel("Comments")
    plt.grid(True)
    plt.show()

# Run Analysis Functions
engagement_over_time(df)
follower_growth(df)
engagement_distribution(df)
correlation_heatmap(df)
engagement_correlation(df)
```
